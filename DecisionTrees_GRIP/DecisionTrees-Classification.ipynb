# Decision Trees - Classification

### Task: Use Decision Trees to visualize iris dataset graphically

### Importing Libraries

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
```

### Creating the DataFrame

```python
iris_data = load_iris()
df = pd.DataFrame(data = iris_data.data, columns=iris_data.feature_names)
y = iris_data.target
df.head()
print(y)
```

### Creating the Classifier

```python
clf = DecisionTreeClassifier()
clf.fit(df,y)
print('Training Completed')
```

### Visualizing the Decision Tree

```python
newfile = StringIO()
data = export_graphviz(clf, out_file=newfile, feature_names= iris_data.feature_names,
                       class_names=iris_data.target_names,filled=True,rounded=True,special_characters=True)
graph = pydotplus.graph_from_dot_data(newfile.getvalue())  
Image(graph.create_png())
```

### Testing against custom data
Try changing the values of the elements of the array

```python
prob = clf.predict_proba(np.array([6,2.3,1.5,1.5]).reshape(1,-1))
print(prob)
```

Here, as we can see, the first, second and third positions indicate the Probablity of a flower sample being a part of the Setosa, Versicolor and Viriginica Species
